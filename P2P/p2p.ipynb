```python
import numpy as np
import pandas as pd
import geopandas
import matplotlib
import matplotlib.pyplot as plt 
import shapely
import csv
```

```python
with open('Pine_to_Palm.csv', 'r') as file:
    reader = csv.reader(file)
    header =  next(reader, None)
    data = []
    for item in reader:
        data.append(item)
    
print(f'{header=}')
print(f'{data=}')
```

```python
distance = []
elevation = []
elev_change = []
slope = []

for item in data:
    distance.append(item[4])
    elevation.append(item[6])
    slope.append(item[7])

elevation = list(map(float, elevation)) 
distance = list(map(float, distance))  
slope = list(map(float, slope))
 
for i in range(len(elevation)):
    change = elevation[i] - elevation[i-1] 
    elev_change.append(change)
   
# print(distance[:5])
# print(elevation[:5])
# print(elev_change[:5])
# print(slope[:5])
```
