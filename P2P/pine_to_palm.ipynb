```python
import numpy as np
import pandas as pd
import geopandas
import matplotlib
import matplotlib.pyplot as plt 
import shapely
```

```python
p2p_complete = pd.read_csv('Pine_to_Palm.csv')
p2p_complete = pd.DataFrame(p2p_complete)
p2p_complete.head()
```

```python
p2p_dist = pd.DataFrame()
p2p_dist['Distance (feet)'] = p2p_complete['Distance (feet)']
p2p_dist['Distance (miles)'] = round(p2p_complete['Distance (miles)'],2)
p2p_dist['Elevation (feet)'] = p2p_complete['Elevation (feet)']
p2p_dist['Slope (degrees)'] = p2p_complete['Slope (degrees)']
p2p_dist.head()
```

```python
p2p_dist.plot('Distance (miles)', 'Elevation (feet)')
```

```python
# p2p_dist['crew'] = [28.5, 41, 43, 52, 66, 80, 100]
crew_miles = [0.0, 28.5, 41.0, 43.0, 52.0, 66.0, 80.0, 100.0]
crew = []
for i in p2p_dist['Distance (miles)']:
    for j in crew_miles:
        if round(i,1) == j:
           crew.append(i)
        else:
            crew.append('NaN') 

crew = pd.DataFrame(crew)            
p2p_dist['crew'] = crew
p2p_dist
```
